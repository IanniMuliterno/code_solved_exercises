```python
import torch
import numpy as np
import matplotlib.pyplot as mtlib
```

```python
x = 2*np.random.rand(100,1)
y = -4 + 3*x + np.random.rand(100,1)

mtlib.scatter(x,y)
mtlib.xlabel('x')
mtlib.ylabel('y')
mtlib.show()
```

```python
mtlib.plot(x)
```

```python
mtlib.plot(y)
```

```python
X_b = np.c_[np.ones((100,1)),x]
# Xb dot product or X*X
dot_product = X_b.T.dot(X_b)
# now lets do the entire formula ((X*X)^(-1))*(X*y)
theta_best = np.linalg.inv(dot_product).dot(X_b.T).dot(y)
```

```python
theta_best
```

### Check goodness of fit

we can now check how good the estimation was

```python
y_hat = theta_best[0] + theta_best[1]*x
mtlib.plot((y_hat - y))
mtlib.title("residual")
#note that they are randomly distributed around 0, as it should be
```

```python
# Define the learning rate and the number of iterations
learning_rate = 0.1
n_iterations = 1000
m = 100

# Initialize theta randomly
theta = np.random.randn(2, 1)

# Add a column of ones to include the intercept term
X_b = np.c_[np.ones((100, 1)), x]

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients

print("Theta (using gradient descent):", theta)
```

```python
# Plot the data and the best fit line
mtlib.scatter(x, y)
# line created from the gradient descent
mtlib.plot(x, X_b.dot(theta), color='red')
mtlib.xlabel('X')
mtlib.ylabel('y')
mtlib.show()
```
